# Get Tags & Frequencies

## 1. Create a list of all tags {'tag' : frequency}

Extract tags from profiles data files and create a list of tags and frequencies.

```python
import json
import os
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Profile data files containing tags
base_dir = os.path.abspath("..")  # go one level back to root
general_profiles = os.path.join(base_dir, "data", "scraped_data", "general_profiles_data.json")
profiles_by_comm = os.path.join(base_dir, "data", "scraped_data", "profiles_by_comm_data.json")
output_path = os.path.join(base_dir, "data", "tag_clustering", "all_tags.json")

all_tags = []

with open(general_profiles, "r", encoding="utf-8") as file:
    profiles = json.load(file)

    for profile in profiles.values():
        all_tags.extend(profile.get("tags", []))

with open(profiles_by_comm, "r", encoding="utf-8") as file:
    profiles_by_comm = json.load(file)

    for community in profiles_by_comm.values():
        for profile in community.values():
            all_tags.extend(profile.get("tags", []))

# Remove 'N/A' tag values
all_tags = [tag for tag in all_tags if tag != "N/A"]

# Count tag frequencies and save to JSON
tag_counts = dict(Counter(all_tags))

with open(output_path, "w", encoding="utf-8") as out:
    json.dump(tag_counts, out, indent=4)

print(f"Total unique tags (excluding 'N/A'): {len(tag_counts)}")
print(f"Saved to: {output_path}")
```

## 2. Filter Tags based on Frequency

### 2.1 Distribtuion of Tag Frequencies

```python
# Get list of frequencies
frequencies = list(tag_counts.values())

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(frequencies, bins=50, edgecolor='black', log=True) # log=True (logarithmic scale) if many rare tags
plt.title("Distribution of Tag Frequencies")
plt.xlabel("Frequency")
plt.ylabel("Number of Tags (log scale)")
plt.grid(True)
plt.show()
```

## Delete Extra Tags from Clustered JSON

```python
import os
import json

# === File paths ===
base_dir = os.path.abspath("..")
tags_original_path = os.path.join(base_dir, "data", "tag_clustering", "all_tags.json")
tags_clustered_path = os.path.join(base_dir, "data", "tag_clustering", "all_tags_filtered_clustered_ChatGPT_o1.json")
tags_cleaned_path = os.path.join(base_dir, "data", "tag_clustering", "tags_clustered_ChatGPT_no_extras.json")

# === Load data ===
with open(tags_original_path, "r") as f_orig:
    all_tags_dict = json.load(f_orig)
with open(tags_clustered_path, "r") as f_clustered:
    clusters_dict = json.load(f_clustered)

# === Step 1: Check for duplicates within each cluster ===
duplicates_found = False
for cluster, tags in clusters_dict.items():
    if len(tags) != len(set(tags)):
        duplicates_found = True
        dupes = [t for t in set(tags) if tags.count(t) > 1]
        print(f"Duplicates in cluster '{cluster}': {dupes}")
if not duplicates_found:
    print("✅ No duplicate tags within clusters.")

# === Step 2: Identify missing tags ===
original_tags = set(all_tags_dict.keys())
clustered_tags = {tag for tag_list in clusters_dict.values() for tag in tag_list}
missing_tags = clustered_tags - original_tags

print(f"\nOriginal tags: {len(original_tags)}")
print(f"Clustered tags: {len(clustered_tags)}")
print(f"Missing tags: {len(missing_tags)}")

if missing_tags:
    print("⚠️ Tags in clustered file but NOT in original:")
    for tag in sorted(missing_tags):
        print(f" - {tag}")
else:
    print("✅ All clustered tags exist in the original file.")

# === Step 3: Remove missing/extraneous tags from clusters ===
for cluster, tag_list in clusters_dict.items():
    clusters_dict[cluster] = [tag for tag in tag_list if tag in original_tags]

# === Step 4: Save cleaned cluster file ===
with open(tags_cleaned_path, "w") as out_f:
    json.dump(clusters_dict, out_f, indent=2)

# === Final stats ===
cleaned_tag_count = sum(len(tags) for tags in clusters_dict.values())
print(f"\n✅ Total tags in cleaned clustered file: {cleaned_tag_count}")
print(f"\n✅ Cleaned cluster file saved to: {tags_cleaned_path}")
```

## Identify Still Unclustered Tags from Origin

## Split Unclustered Tags into Bacthes ~300

```python
import os
import json
from math import ceil

base_dir = os.path.abspath("..")
input_path = os.path.join(base_dir, "data", "tag_clustering", "unclustered_tags.json")
output_dir = os.path.join(base_dir, "data", "tag_clustering", "unclustered")

with open(input_path, "r", encoding="utf-8") as f:
    tags = json.load(f)

# === Parameters ===
batch_size = 300
num_batches = ceil(len(tags) / batch_size)

# === Split and save ===
for i in range(num_batches):
    start = i * batch_size
    end = start + batch_size
    batch_tags = tags[start:end]
    
    batch_file = os.path.join(output_dir, f"unclustered_tags_batch_{i+1}.json")
    with open(batch_file, "w", encoding="utf-8") as f:
        json.dump(batch_tags, f, indent=2)
    
    print(f"✅ Saved batch {i+1} with {len(batch_tags)} tags to: {batch_file}")
```
